## How are in-demand skills trending for Data Analysts?

```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  
import ast
import seaborn as sns
from matplotlib.ticker import PercentFormatter

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)

df_DA = df[df['job_title_short'] == 'Data Analyst'].copy()
```

```python
df_DA_US = df_DA[df_DA['job_country'] == 'United States'].copy()
df_DA_US
```

```python
df_DA_US['job_posted_month_no'] = df_DA_US['job_posted_date'].dt.month
df_DA_US
```

```python
df_DA_US_explode = df_DA_US.explode('job_skills').copy()
df_DA_US_explode
```

```python
df_DA_US_pivot = df_DA_US_explode.pivot_table(index='job_posted_month_no', columns='job_skills', aggfunc='size', fill_value=0)
df_DA_US_pivot.loc['total'] = df_DA_US_pivot.sum()
df_DA_US_pivot = df_DA_US_pivot[df_DA_US_pivot.loc['total'].sort_values(ascending=False).index]
df_DA_US_pivot = df_DA_US_pivot.drop('total')
df_DA_US_pivot
```

```python
# Get monthly totals
DA_totals = df_DA_US.groupby('job_posted_month_no').size()
df_DA_per = df_DA_US_pivot.div(DA_totals/100, axis=0)
df_DA_per
```

```python
df_DA_per = df_DA_per.reset_index()
df_DA_per
```

```python
df_DA_per['job_posted_month'] = df_DA_per['job_posted_month_no'].apply(lambda x: pd.to_datetime(x, format='%m').strftime('%b'))
df_DA_per
```

```python
#df_DA_per = df_DA_per.set_index('job_posted_month')
df_DA_per  = df_DA_per.drop(columns='job_posted_month_no')
df_DA_per
```

```python
df_plot = df_DA_per.iloc[:,:15]
sns.lineplot(data=df_plot, dashes=False, legend='full', palette='tab10')
sns.set_theme(style='ticks')
sns.despine()


plt.title('Trending Top Skills for Data Analyst in the US')
plt.ylabel('Likelihood in Job Postings %')
plt.xlabel('2023')
plt.legend().remove()
plt.gca().yaxis.set_major_formatter(PercentFormatter(decimals=0))

# annotate the plot with the top 5 skills using plt.text()
for i in range(15):
    plt.text(11.2, df_plot.iloc[-1, i], df_plot.columns[i], color='black')

plt.show()
plt.tight_layout()
```
